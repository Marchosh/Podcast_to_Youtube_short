```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# Create an empty DataFrame with columns
df = pd.DataFrame(columns=["podcastID", "transcript"])

# URL of the website
url = "https://www.biggerpockets.com/podcast-transcripts"

# Send a GET request to the URL and retrieve the HTML content
response = requests.get(url)
html_content = response.content

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(html_content, "html.parser")

# Find the <div> element with the ID "primary"
primary_div = soup.find("div", {"id": "primary"})

# Find the second <p> tag within the <div> element
second_p_tag = primary_div.find_all("p")[1]

# Find all the <a> tags within the second <p> tag
a_tags = second_p_tag.find_all("a")

# Iterate through the <a> tags and extract the links
for a_tag in a_tags:
    # Get the URL of the linked website
    link = a_tag.get("href")
    
    # Check if the link is valid
    try:
        response = requests.get(link)
        response.raise_for_status()  # Raise an exception for non-200 status codes
    except requests.exceptions.RequestException:
        print(f"Skipping invalid link: {link}")
        continue

    # URL of the news article
    url = link

    # Send a GET request to the URL and retrieve the HTML content
    response = requests.get(url)
    html_content = response.content

    # Create a BeautifulSoup object to parse the HTML
    soup = BeautifulSoup(html_content, "html.parser")

    # Find the <div> element with the ID "primary"
    primary_div = soup.find("div", {"id": "primary"})

    # Extract the plain text from the <div> element
    article_text = primary_div.get_text(separator=" ")
    
    #Set an ID
    # Extract the text and number from the URL using regular expressions
    match = re.search(r"bpp(\d+)", url)
    if match:
        podcastID = match.group(0)

    # Create a DataFrame with the new article
    new_article = pd.DataFrame([[podcastID, article_text]], columns=["podcastID", "transcript"])

    # Append the new article to the existing DataFrame
    df = df.append(new_article, ignore_index=True)
```
