```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-3.5 model and tokenizer
model_name = "EleutherAI/gpt-3.5-turbo"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Sample YouTube transcript
youtube_transcript = """
[00:00:00] Host: Welcome back to our channel, folks.
[00:00:05] Guest 1: Hi, everyone! Today, we're going to talk about the latest tech trends.
[00:00:12] Host: Absolutely, and there's one technology that's been making waves lately...
[00:00:18] Guest 2: That's right, artificial intelligence. It's transforming industries worldwide.
[00:00:25] Host: And we have a special guest with us who's an expert in this field.
[00:00:30] Guest 3: Thank you for having me. AI is indeed reshaping how we live and work.
[00:00:37] Host: Fascinating! Can you give us some examples of AI applications in daily life?
[00:00:43] Guest 3: Of course. AI powers virtual assistants, recommendation systems, and even self-driving cars.
[00:00:51] Guest 1: That's incredible. I'm particularly amazed by self-driving cars.
[00:00:57] Host: It's impressive how AI algorithms can navigate and make split-second decisions.
[00:01:04] Guest 2: Absolutely, and the technology keeps advancing, opening up new possibilities.
[00:01:11] Host: Thank you for sharing your insights. Stay tuned for more tech updates!
"""

# Tokenize the transcript and generate keywords
input_ids = tokenizer.encode(youtube_transcript, return_tensors="pt", max_length=1024, truncation=True)
with torch.no_grad():
    outputs = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.92)

# Decode and print the generated text (potential interesting word or phrase)
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Text:", generated_text)
```
